```python
# 5.   Diferencie variável local e global em funções.

my_number = 1
print(f"Número my_number inicial: {my_number}")

def change_number():
  my_number = 2
  print(f"{my_number} -- Dentro da função, não altera a my_number de fora")


def super_change_number():
  global my_number
  my_number = 3


# Local. Dentro da função, a variável de mesmo nome não altera ou não acessa a variável de mesmo nome criada fora da função
change_number()
print(f"{my_number} -- Fora da função, não exibe o valor alterado dentro da função. Mantém o valor global")

# Global.
super_change_number()
print(f"{my_number} -- De forma global, agora a alteração da função afeta a variável global")
```

```python
# 23.  Implemente um algoritmo que remova valores duplicados de uma lista.

the_list = [1, 2, 3, 3]

def remove_duplicated(the_list):
  new_list = []
  for i in the_list:
    if i not in new_list:
      new_list.append(i)

  return new_list


remove_duplicated(the_list)
```

```python
# 27.  Inverta uma string sem usar funções prontas.

def invert_string(text):
  inverted_string = ""
  my_list = []

  for i in text:
    my_list.append(i)

  for i in range(len(my_list) -1, -1, -1):
    inverted_string += my_list[i]

  return inverted_string


print(invert_string("Data Science Senai"))

# Aprendemos usar []. Dá para inverter:
print(invert_string("Data Science Senai")[::-1])

```

```python
# 32.  Implemente uma função que calcule o fatorial de um número.

# vou usar o modo Recursivo que a Professora Aruane passou na penúltima aula

def factorial(number):
    if number == 0 or number == 1:
        return 1
    return number * factorial(number - 1)


user_choice = int(input("Type the number: "))
print(factorial(user_choice))
```

```python
# 34.  Desenvolva uma função que verifique se um número é primo.

def check_prime_number(num):
  prime = False if num <= 1 else True

  for i in range(2, num):
    if num % i == 0:
      prime = False
      break

  return prime


check_prime_number(13)
```
