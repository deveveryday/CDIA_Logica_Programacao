#### 12) Calcule a taxa de letalidade por país

```python
# total cases / cases by country
df_letality = df.groupby("country")[["total_deaths", "total_cases"]].max()
df_letality["letality"] = ((df_letality["total_deaths"] / df_letality["total_cases"]) * 100).round(2)

df_letality.sort_values(by = "letality", ascending = False).head(30)
```

```python

# Configuração inicial
plt.style.use('seaborn-v0_8')
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'

# 1. Cabeçalho do Dashboard
plt.figure(figsize=(14, 1))
plt.text(0.5, 0.5, 'Dashboard - Análise do Conjunto de dados Iris',
         ha='center', va='center',
         fontsize=24, fontweight='bold', color='#2e86ab')
plt.axis('off')
plt.show()

# 2. Visualização Geral (Subplots)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Principais Análises', y=1.02, fontsize=18, fontweight='bold')

# Gráfico 1: Histograma
sns.histplot(data=df_letality, x="country", hue="total_deaths",
             kde=True, ax=axes[0, 0], palette="husl")
axes[0, 0].set_title('Distribuição do Comprimento da Sépala')
```

```python

# Gráfico 2: Boxplot
sns.boxplot(data=df_letality, x="country", y="total_deaths",
            hue="country", palette="husl", legend=False, ax=axes[0, 1])
axes[0, 1].set_title('Largura da Pétala por Espécie')

# Gráfico 3: Scatter Plot
sns.scatterplot(data=df_letality, x="country", y="total_deaths",
                hue="species", style="species", s=100,
                palette="husl", ax=axes[1, 0])
axes[1, 0].set_title('Relação Sépala vs Pétala')

# Gráfico 4: Heatmap de Correlação
corr = df_letality.iloc[:, :4].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm",
            center=0, ax=axes[1, 1], fmt=".2f")
axes[1, 1].set_title('Matriz de Correlação')

plt.tight_layout()
plt.show()

# 3. Gráfico Adicional (Pair Plot)
plt.figure(figsize=(12, 8))
sns.pairplot(df_letality, hue="country", palette="husl",
             height=2.5, plot_kws={'s': 80})
plt.suptitle('Relações entre Todas as Variáveis', y=1.02)
plt.show()

# 4. Estatísticas Resumidas
print("\n Métricas Estatísticas por Espécie:")
stats = df_letality.groupby('country').agg(['mean', 'std', 'min', 'max']).round(2)
display(stats.style.background_gradient(cmap='Blues', axis=0))
```
