### Passo 4: Anonimização dos dados (opcional)

A **Lei Geral de Proteção de Dados (LGPD - Lei nº 13.709/2018)** é a legislação brasileira que regula o tratamento de dados pessoais, assegurando direitos aos titulares e impondo obrigações a empresas e órgãos públicos. Seu objetivo é proteger a privacidade e garantir a segurança das informações pessoais.

#### O que é dado pessoal?
É qualquer informação relacionada a uma pessoa identificada ou identificável, como:

- Nome, CPF, RG, e-mail, telefone, endereço.

- Dados de localização, IP, cookies, hábitos de consumo.

#### O que é dado sensível?
São dados pessoais que podem causar discriminação ou danos se vazados, como:

- Origem racial ou étnica, convicções religiosas, opinião política.

- Dados biométricos, de saúde, vida sexual, filiação sindical.

- A LGPD exige **proteção reforçada** para esses dados.

#### Importância da anonimização
A **anonimização** é o processo que remove ou altera dados para que uma pessoa não possa ser identificada.

É importante porque:

- **Reduz riscos** de vazamentos e uso indevido.

- Permite o uso de dados para **pesquisas e estatísticas** sem violar a privacidade.

- Dados anonimizados **não são considerados pessoais**, ficando fora do escopo da LGPD

A anonimização de dados é essencial para proteger dados pessoais e/ou quaisquer referências que possam identificar

e sensíveis (como CPF, nomes, e-mails) antes de compartilhar bases de dados.
A seguir, vamos explorar alguns métodos usando Pandas e bibliotecas auxiliares:

#### 4.1. Pseudonimização reversível (Criptografia)
Use chaves de criptografia para transformar dados em pseudônimos e depois recuperá-los quando autorizado.

````python
#!pip install cryptography

from cryptography.fernet import Fernet

# Generate a Key
key = Fernet.generate_key()
key
````

````python
cipher = Fernet(key)

def crypt(text):
  if pd.isna(text):
    return None
  return cipher.encrypt(str(text).encode()).decode()

def decrypt(pseudonym_text):
  if pd.isna(pseudonym_text):
    return None
  return cipher.decrypt(str(pseudonym_text).encode()).decode()


df["Name"] = df["Name"].apply(crypt)
df.head()

````

````python

df["Name"] = df["Name"].apply(decrypt)
df.head()
````

````python
names = [1]
print(df[df["PassengerId"].isin(names)])
````

#### 4.2. Pseudonimização irreversível (Hash de Identificadores)
Transforma dados sensíveis em hashes irreversíveis (útil para manter referências únicas sem expor dados reais).

Se você **não precisa reverter**, mas quer garantir que o mesmo dado gere sempre o mesmo pseudônimo, use **hash com salt**.

````python
import hashlib

# Salt (a secret value by user)
MY_SALT = "mina do cabelo preto"

def pseudominimize(text, salt = MY_SALT):
  return hashlib.sha256((text + salt).encode()).hexdigest()[:10]

df["Name"] = df["Name"].apply(pseudominimize)
````

````python
df.head()
````

#### Como Vincular Dados Futuros?
- Sempre que o mesmo dado aparecer, ele gerará o mesmo hash.
- Por exemplo, sempre que um mesmo CPF aparecer em uma entrada futura.
- Você pode comparar hashes para vincular registros sem saber o dado original.

````python
new_name = "Braund, Mr. Owen Harris"
new_pseudonyn_name = pseudominimize(new_name)
print(df["Name"][0], new_pseudonyn_name)
if new_name == df["Name"][0]:
  print("same person")
````

#### Quando Usar?
- Em bancos de dados analíticos onde você não precisa reverter.

- Para evitar duplicações (ex: contar usuários únicos sem saber quem são).
