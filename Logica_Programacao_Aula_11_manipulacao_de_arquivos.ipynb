## 1. Manipulação de Arquivos

Em muitas aplicações, os dados não ficam apenas na memória do programa. Eles precisam ser **armazenados** para uso posterior.

Alguns exemplos:

- salvar relatórios;
- registrar logs de execução;
- armazenar resultados de análises;
- ler bases de dados simples em formato texto;
- trabalhar com imagens, áudios, PDFs e outros arquivos binários.

Na lógica da programação, manipular arquivos é importante porque permite criar programas que **persistem dados**, ou seja, que guardam informação mesmo após o encerramento da execução.

## 2. Conceitos iniciais

Em Python, o trabalho com arquivos normalmente envolve cinco ações fundamentais:

1. **Criação** do arquivo.
2. **Abertura** do arquivo.
3. **Leitura** do conteúdo.
4. **Escrita** de novos dados.
5. **Fechamento** do arquivo.

A função principal usada para isso é `open()`.

A estrutura básica é:

```python
arquivo = open("nome_do_arquivo.txt", "modo")
```

O segundo parâmetro indica o **modo de abertura**.


## 3. Modos de abertura mais usados

| Modo | Significado |
|---|---|
| `r` | leitura |
| `w` | escrita, cria o arquivo ou sobrescreve se já existir |
| `a` | adiciona conteúdo ao final do arquivo |
| `b` | modo binário |
| `t` | modo texto, padrão do Python |
| `x` | criação exclusiva, gera erro se o arquivo já existir |

Exemplos:

- `"r"` → leitura de arquivo texto
- `"w"` → escrita de arquivo texto
- `"rb"` → leitura de arquivo binário
- `"wb"` → escrita de arquivo binário

## 5. Criação de arquivo de texto

Um arquivo de texto pode ser criado com o modo `w` ou `x`.

- `w` cria o arquivo se ele não existir;
- `w` apaga o conteúdo anterior se o arquivo já existir;
- `x` cria somente se o arquivo ainda não existir.

Existem várias formas de leitura:

- `read()` lê todo o conteúdo;
- `readline()` lê uma linha por vez;
- `readlines()` lê todas as linhas e devolve uma lista.

## 9. Fechamento de arquivos de texto

Fechar o arquivo é uma etapa importante, pois garante:

- liberação de recursos do sistema;
- gravação correta dos dados pendentes;
- redução de erros durante novas operações.

Embora seja possível usar `close()`, em Python é mais seguro utilizar a estrutura `with open(...) as ...`, pois ela fecha o arquivo automaticamente.

O with é uma estrutura do Python usada para **gerenciar recursos automaticamente**, garantindo que eles sejam **abertos e fechados corretamente.**

## 15. Comparação entre arquivo texto e arquivo binário

| Característica | Arquivo de texto | Arquivo binário |
|---|---|---|
| Conteúdo | caracteres legíveis | bytes |
| Modos comuns | `r`, `w`, `a` | `rb`, `wb`, `ab` |
| Leitura humana direta | sim | geralmente não |
| Exemplos | `.txt`, `.csv`, `.json` | `.jpg`, `.png`, `.pdf`, `.mp3` |
