## 5. CriaĂ§ĂŁo de arquivo de texto

Um arquivo de texto pode ser criado com o modo `w` ou `x`.

- `w` cria o arquivo se ele nĂŁo existir;
- `w` apaga o conteĂşdo anterior se o arquivo jĂˇ existir;
- `x` cria somente se o arquivo ainda nĂŁo existir.

No exemplo abaixo, vamos criar um arquivo de texto usando `w`.


````python
from pathlib import Path

folder = Path("files_class")
folder.mkdir(exist_ok = True)

print(f"Work Folder created in: {folder.resolve()}")
````

````python
file_text = folder / "sample_text.txt"

file = open(file_text, "w", encoding = "utf-8")

file.write("first line")

file.close()
````

````python
def mergeSort(arr):
  if len(arr) <= 1:
    return arr

  mid = len(arr) // 2
  leftHalf = arr[:mid]
  rightHalf = arr[mid:]

  sortedLeft = mergeSort(leftHalf)
  sortedRight = mergeSort(rightHalf)

  return merge(sortedLeft, sortedRight)

def merge(left, right):
  result = []
  i = j = 0

  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1

  result.extend(left[i:])
  result.extend(right[j:])

  return result
````

## 6. Abertura e leitura de arquivo de texto

Depois de criar o arquivo, podemos abri-lo no modo `r` para leitura.

Existem vĂˇrias formas de leitura:

````python
file = open(file_text, "r", encoding = "utf-8")

content = file.read()
file.close()
content
````

## 9. Fechamento de arquivos de texto

Fechar o arquivo Ă© uma etapa importante, pois garante:

- liberaĂ§ĂŁo de recursos do sistema;
- gravaĂ§ĂŁo correta dos dados pendentes;
- reduĂ§ĂŁo de erros durante novas operaĂ§Ăµes.

Embora seja possĂ­vel usar `close()`, em Python Ă© mais seguro utilizar a estrutura `with open(...) as ...`, pois ela fecha o arquivo automaticamente.

O with Ă© uma estrutura do Python usada para **gerenciar recursos automaticamente**, garantindo que eles sejam **abertos e fechados corretamente.**

````python
with open(file_text, "r", encoding = "utf-8") as file:
  content_with = file.read()

print(content_with)
````

## 10. Exemplo prĂˇtico com lista de alunos

Vamos criar um pequeno exemplo mais prĂłximo da rotina de anĂˇlise de dados. Imagine que desejamos registrar nomes de alunos em um arquivo texto.


````python
file_students = folder / "students.txt"

students = ["Gabe", "Janice", "Roberta", "Mia", "Sheron"]

with open(file_students, "w", encoding = "utf-8") as file:
  for i in students:
    file.write(i + "\n")
````

````python
with open(file_students, "r", encoding = "utf-8") as file:
  for i in file:
    print(i.strip())
````

# Parte 2, Arquivos binĂˇrios

Arquivos binĂˇrios armazenam dados em bytes. Eles nĂŁo sĂŁo, em geral, legĂ­veis diretamente por humanos.

Exemplos:

- imagens, como `.png` e `.jpg`;
- Ăˇudios, como `.mp3`;
- vĂ­deos, como `.mp4`;
- arquivos compactados, como `.zip`;
- arquivos serializados e formatos proprietĂˇrios.

Ao trabalhar com arquivos binĂˇrios em Python, usamos os modos com `b`, como `rb` e `wb`.


## 11. CriaĂ§ĂŁo de arquivo binĂˇrio

Para criar um arquivo binĂˇrio, podemos escrever uma sequĂŞncia de bytes. No exemplo a seguir, vamos criar um arquivo binĂˇrio simples com valores numĂ©ricos.


````python
binary_file = folder / "data.bin"

data = bytes([10, 20, 30, 40, 50])

file = open(binary_file, "wb")

file.write(data)
file.close()

print(binary_file)
#
````

## 12. Abertura e leitura de arquivo binĂˇrio

Agora vamos abrir o arquivo binĂˇrio em modo `rb` e ler seu conteĂşdo.

Perceba que o resultado serĂˇ exibido em bytes.


````python
file = open(binary_file, "rb")

binary_content = file.read()

file.close()

print(list(binary_content))
````

## 13. Escrita adicional em arquivo binĂˇrio

TambĂ©m Ă© possĂ­vel acrescentar dados ao final de um arquivo binĂˇrio usando `ab`.


````python
file = open(binary_file, "ab")

file.write(bytes([60, 70, 80, 90]))
file.close()
````

````python
with open(binary_file, "rb") as file:
  updated_data = file.read()

print(updated_data)
print(list(updated_data))
````
