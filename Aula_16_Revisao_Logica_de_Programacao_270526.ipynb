# 1. Fundamentos de lógica de programação

Lógica de programação é o conjunto de princípios utilizados para organizar instruções de forma coerente, finita e não ambígua, com o objetivo de resolver um problema computacional.

Um programa pode ser entendido como uma implementação de um algoritmo em uma linguagem de programação.

## Algoritmo

Um algoritmo é uma sequência finita de passos bem definidos que transforma dados de entrada em resultados de saída.

$
\text{Entrada} \rightarrow \text{Processamento} \rightarrow \text{Saída}
$

## Exemplo informal

Problema: calcular a média de três notas.

Passos:

1. receber três notas;
2. somar as três notas;
3. dividir a soma por três;
4. exibir o resultado.

````python
# Exemplo inicial: cálculo da média de três notas

nota1 = 7.0
nota2 = 8.5
nota3 = 9.0

media = (nota1 + nota2 + nota3) / 3

print("Média:", media)
````

# 2. Tipos de erro em programação

Durante o desenvolvimento de programas, é comum encontrar diferentes classes de erro.

## 2.1 Erro de sintaxe

Ocorre quando o código viola as regras formais da linguagem. O interpretador Python normalmente identifica esse erro antes ou durante a execução.

Exemplo conceitual:

```python
if idade >= 18
    print("Maior de idade")
```

O erro ocorre porque falta `:` ao final da condição.

## 2.2 Erro de execução

Ocorre quando o programa possui sintaxe válida, mas falha durante a execução.

Exemplo:

```python
x = 10 / 0
```

Nesse caso, ocorre uma divisão por zero.

## 2.3 Erro de lógica

Ocorre quando o programa executa sem falhas aparentes, mas produz um resultado incorreto ou incoerente com o problema.

Exemplo:

```python
media = nota1 + nota2 + nota3 / 3
```

Nesse caso, apenas `nota3` é dividida por 3. O correto seria usar parênteses.

````python
# Erro de lógica: média calculada de forma incorreta

nota1 = 7
nota2 = 8
nota3 = 9

media_errada = nota1 + nota2 + nota3 / 3
media_correta = (nota1 + nota2 + nota3) / 3

print("Média errada:", media_errada)
print("Média correta:", media_correta)
````

A precedência dos operadores é uma fonte frequente de erros de lógica. Use parênteses quando a expressão puder ser interpretada de forma ambígua.

# 6. Estrutura condicional `if`, `elif` e `else`

Condicionais permitem executar blocos diferentes dependendo de uma condição.

Estrutura geral:

```python
if condição:
    bloco_se_verdadeiro
elif outra_condição:
    bloco_alternativo
else:
    bloco_final
```

A ordem das condições pode alterar completamente o resultado. Condições mais específicas devem aparecer antes de condições mais gerais.

````python
nota = 6.5

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
````

````python
# Exemplo com ordem incorreta

nota = 8.5

if nota >= 0:
    print("Nota válida")
elif nota >= 7:
    print("Aprovado")

# O bloco "Aprovado" nunca será alcançado,
# pois qualquer nota maior ou igual a 7 também é maior ou igual a 0.
````

````python
# Exemplo com ordem adequada

nota = 8.5

if nota >= 7:
    print("Aprovado")
elif nota >= 0:
    print("Nota válida, mas abaixo da média")
else:
    print("Nota inválida")
````

# 7. Laços de repetição

Laços permitem repetir instruções.

Python possui duas estruturas principais:

1. `for`: usado quando se deseja percorrer uma sequência ou repetir um número conhecido de vezes;
2. `while`: usado quando a repetição depende de uma condição.

## Laço `for`

O laço `for` percorre elementos de uma sequência.

## Laço `while`

O laço `while` repete enquanto uma condição for verdadeira. É importante garantir que a condição se torne falsa em algum momento.

````python
nomes = ["Alice", "Bruno", "Carlos"]

for nome in nomes:
    print("Nome:", nome)

print("\nUsando range:")

for i in range(5):
    print(i)
````

````python
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1
````

# 8. Contadores e acumuladores

Contadores e acumuladores são padrões fundamentais de lógica de programação.

## Contador

Registra quantas vezes algo ocorreu.

## Acumulador

Soma ou agrega valores ao longo de uma repetição.

````python
# Contador de números pares

contador_pares = 0

for numero in range(1, 11):
    if numero % 2 == 0:
        contador_pares += 1

print("Quantidade de pares:", contador_pares)
````

````python
# Acumulador de soma

soma = 0

for numero in range(1, 11):
    soma += numero

print("Soma de 1 a 10:", soma)
````

````python
# Erro comum: sobrescrever em vez de acumular

soma = 0

for numero in range(1, 6):
    soma = numero  # erro lógico: sobrescreve a soma

print("Resultado incorreto:", soma)

soma = 0

for numero in range(1, 6):
    soma += numero  # correto: acumula os valores

print("Resultado correto:", soma)
````

## Erro sutil: remover elementos enquanto percorre a lista

Modificar uma lista durante sua própria iteração pode produzir resultados inesperados.

````python
# Exemplo problemático

valores = [1, 2, 3, 4, 6, 8, 9, 10]

for valor in valores:
    if valor % 2 == 0:
        valores.remove(valor)

print(valores)

# Dependendo da posição dos elementos, alguns itens podem ser ignorados.
````

````python
# Forma recomendada: criar uma nova lista com os valores desejados

valores = [1, 2, 3, 4, 6, 8, 9, 10]

sem_pares = []

for valor in valores:
    if valor % 2 != 0:
        sem_pares.append(valor)

print(sem_pares)
````

# 11. Funções

Funções permitem organizar o código em blocos reutilizáveis.

Estrutura geral:

```python
def nome_da_funcao(parametros):
    processamento
    return resultado
```

Funções são importantes para reduzir repetição, facilitar testes e melhorar a clareza do programa.

````python
def calcular_media(notas):
    """
    Calcula a média aritmética de uma lista de notas.
    """
    soma = sum(notas)
    quantidade = len(notas)
    media = soma / quantidade
    return media


def classificar_aluno(media):
    """
    Classifica o aluno com base na média.
    """
    if media >= 7:
        return "Aprovado"
    elif media >= 5:
        return "Recuperação"
    else:
        return "Reprovado"


notas = [7, 8, 9]
media = calcular_media(notas)
situacao = classificar_aluno(media)

print("Média:", media)
print("Situação:", situacao)
````

# 13. Erros de lógica difíceis de perceber

A seguir são apresentados exemplos de erros que frequentemente passam despercebidos porque o código executa normalmente.

O foco é analisar o comportamento produzido, não apenas verificar se há mensagens de erro.

## 13.1 Mutabilidade compartilhada em listas aninhadas

### Código com erro lógico

````python
matriz = [[0] * 3] * 3

matriz[0][0] = 99

print(matriz)
````

### Correção

````python
matriz = [[0 for _ in range(3)] for _ in range(3)]

matriz[0][0] = 99

print(matriz)
````

### Explicação

O operador de repetição cria referências para a mesma sublista. Ao alterar uma linha, todas as linhas são afetadas.

## 13.2 Argumento padrão mutável

### Código com erro lógico

````python
def adicionar_item(item, lista=[]):
    lista.append(item)
    return lista

print(adicionar_item("A"))
print(adicionar_item("B"))
````

### Correção

````python
def adicionar_item(item, lista=None):
    if lista is None:
        lista = []
    lista.append(item)
    return lista

print(adicionar_item("A"))
print(adicionar_item("B"))
````

### Explicação

Argumentos padrão são avaliados uma única vez. A lista persiste entre chamadas da função.

## 13.3 Comparação de floats por igualdade

### Código com erro lógico

````python
x = 0.1 + 0.2

if x == 0.3:
    print("Igual")
else:
    print("Diferente")

print(x)
````

### Correção

````python
x = 0.1 + 0.2

if abs(x - 0.3) < 1e-9:
    print("Aproximadamente igual")
else:
    print("Diferente")

print(x)
````

### Explicação

Números reais são representados em base binária, o que pode introduzir pequenas diferenças numéricas.

## 13.6 Cópia rasa de lista aninhada

### Código com erro lógico

````python
original = [[1, 2], [3, 4]]
copia = original.copy()

copia[0][0] = 99

print("Original:", original)
print("Cópia:", copia)
````

### Correção

````python
import copy

original = [[1, 2], [3, 4]]
copia = copy.deepcopy(original)

copia[0][0] = 99

print("Original:", original)
print("Cópia:", copia)
````

### Explicação

A cópia rasa duplica apenas a estrutura externa. As listas internas continuam compartilhadas.

## 13.12 Valor inicial inadequado para encontrar máximo

````python
numeros = [-10, -3, -7]

maior = 0

for numero in numeros:
    if numero > maior:
        maior = numero

print(maior)
````

# 14. Depuração: como encontrar erros de lógica

A depuração é o processo de investigar o comportamento do código para localizar falhas.

Estratégias recomendadas:

1. testar casos simples;
2. imprimir valores intermediários;
3. verificar limites de intervalos;
4. comparar saída obtida com saída esperada;
5. usar `assert`;
6. dividir o problema em funções menores;
7. testar casos extremos.

````python
def calcular_desconto(preco, percentual):
    """
    Calcula o preço final após desconto percentual.
    """
    desconto = preco * percentual / 100
    preco_final = preco - desconto
    return preco_final


# Testes simples
print(calcular_desconto(100, 10))  # esperado: 90
print(calcular_desconto(200, 25))  # esperado: 150
print(calcular_desconto(50, 0))    # esperado: 50
````

````python
# Uso de assert para validar resultados

assert calcular_desconto(100, 10) == 90
assert calcular_desconto(200, 25) == 150
assert calcular_desconto(50, 0) == 50

print("Todos os testes passaram.")
````

# 15. Desenvolvimento incremental

Uma forma eficiente de reduzir erros de lógica é construir o programa em etapas.

Exemplo de problema:

> Receber uma lista de notas, calcular a média e classificar o aluno.

Etapas:

1. armazenar as notas;
2. calcular a média;
3. classificar a média;
4. exibir o resultado;
5. testar diferentes casos.

````python
def calcular_media(notas):
    return sum(notas) / len(notas)


def classificar(media):
    if media >= 7:
        return "Aprovado"
    elif media >= 5:
        return "Recuperação"
    else:
        return "Reprovado"


notas = [6, 7, 8]

media = calcular_media(notas)
situacao = classificar(media)

print("Notas:", notas)
print("Média:", media)
print("Situação:", situacao)
````

````python
# Exercício 1
# Escreva um programa que receba três números e informe o maior deles.

def find_bigger(num1, num2, num3):
  if num1 > num2 and num1 > num3:
    return num1
  elif num2 > num1 and num2 > num3:
    return num2
  elif num3 > num1 and num3 > num2:
    return num3


find_bigger(1, 40, 3)
````

````python
# Exercício 2
# Crie um programa que conte quantos números pares existem em uma lista.

def count_list(list_one):
  result = 0

  for i in list_one:
      result += 1 if i % 2 == 0 else 0

  return result

count_list([1, 2, 4])
````

````python
# Exercício 3
# Dada uma lista de notas, calcule a média e indique se o aluno foi aprovado, ficou em recuperação ou foi reprovado.

def calculate_mean(scores_student):
  result = 0

  for i in scores_student:
    result += i

  return result / len(scores_student)

def evaluate_student(mean):
  if mean >= 7:
    return "Approved"
  elif mean >= 5:
    return "Recuperation"
  else:
    return "Reproved"


evaluate_student(calculate_mean([0, 7, 7]))
````

````python
# Exercício 4
# Crie uma função que receba uma palavra e informe se ela é um palíndromo.

def check_palindrome(word):
  return word == word[::-1]


print("Is palidrome:", check_palindrome("ama"))
````

````python
# Exercício 5
# Crie um programa que remova valores duplicados de uma lista, preservando a ordem original.

def remove_duplicates(list_one):
  new_list = []

  for i in list_one:
    if i not in new_list:
      new_list.append(i)

  return new_list


remove_duplicates([1, 1, 2, 2])
````

````python
# Exercício 6
# Crie uma função que receba uma lista de números e retorne o maior valor, sem usar max.

def get_bigger(list_one):
  bigger = list_one[0]

  for i in list_one:
    if i > bigger:
      bigger = i

  return bigger


get_bigger([21, 1, 0])
````

````python
# Exercício 7
# Crie uma função que identifique se um número é primo.

def is_prime(number):
  prime = False if number <= 1 else True

  for i in range(2, int(number ** 0.5) + 1):
    if number % i == 0:
      prime = False
      break

  return prime


is_prime(50)
````

````python
# Exercício 8
# Crie um programa que conte a frequência de cada palavra em uma frase.

def count_words(phrase):
  result = {}

  for word in phrase:
    result[word] = 0

  for word in phrase:
    result[word] = int(result[word]) + 1

  return result


count_words("ju$$$$$tiiino")
````
