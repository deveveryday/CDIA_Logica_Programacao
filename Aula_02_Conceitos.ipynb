### Exemplo: calcular área de um retângulo

**Pseudocódigo**
1. Ler base
2. Ler altura
3. área = base * altura
4. Mostrar área

```python
base = float(input("Type the Base size"))
height = float(input("Type the Height"))

print("The total Area is ", base * height)
```

### Exercício: calcular a média de duas notas
**Pseudocódigo**

1. Ler nota1
2. Ler nota2
3. media = (nota1 + nota2) / 2
4. Mostrar media

```python
import numpy as np #pattern

first_score = float(input("Type your Score in the first test:: "))
second_score = float(input("Type your Score in the second test:: "))

print("Your Score is: [", (first_score + second_score) / 2, "]")
```
