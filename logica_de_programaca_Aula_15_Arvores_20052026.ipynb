## ÁRVORE BINÁRIA DE BUSCA (BST - Binary Search Tree)
### Implementação utilizando dicionários em Python

```python
# ------------------------------------------------------------
# Cria um dicionário vazio que armazenará toda a árvore.
#
# Estrutura:
# {
#     valor_do_no: {
#         "esquerda": valor_filho_esquerdo,
#         "direita": valor_filho_direito
#     }
# }
# ------------------------------------------------------------
arvore = {}


# ------------------------------------------------------------
# Lista de valores que serão inseridos na árvore binária.
# A ordem de inserção influencia diretamente a estrutura final.
# ------------------------------------------------------------
valores = [53, 30, 72, 14, 39, 61, 84, 9, 23, 34, 49, 79]


# ============================================================
# INSERÇÃO DOS ELEMENTOS NA ÁRVORE
# ============================================================

# ------------------------------------------------------------
# Percorre cada valor da lista.
# Cada elemento será inserido respeitando a regra:
#
# menor  -> lado esquerdo
# maior  -> lado direito
# ------------------------------------------------------------
for valor in valores:

    # --------------------------------------------------------
    # Exibe qual valor está sendo inserido.
    # --------------------------------------------------------
    print("\n================================================")
    print(f"Iniciando inserção do valor: {valor}")
    print("================================================")

    # --------------------------------------------------------
    # Verifica se a árvore ainda está vazia.
    # O primeiro elemento inserido será a raiz.
    # --------------------------------------------------------
    if len(arvore) == 0:

        # ----------------------------------------------------
        # Cria o nó raiz.
        #
        # "esquerda": None
        # "direita": None
        #
        # None indica ausência de filhos.
        # ----------------------------------------------------
        arvore[valor] = {
            "esquerda": None,
            "direita": None
        }

        # ----------------------------------------------------
        # Define o valor atual como raiz da árvore.
        # ----------------------------------------------------
        raiz = valor

        # ----------------------------------------------------
        # Exibe informação da criação da raiz.
        # ----------------------------------------------------
        print(f"O valor {valor} foi definido como RAIZ da árvore.")

    else:

        # ----------------------------------------------------
        # Começa a busca a partir da raiz.
        # A variável "atual" representa o nó corrente.
        # ----------------------------------------------------
        atual = raiz

        # ----------------------------------------------------
        # Exibe o ponto inicial da navegação.
        # ----------------------------------------------------
        print(f"Iniciando busca a partir da raiz: {raiz}")

        # ----------------------------------------------------
        # Loop de navegação até encontrar
        # uma posição livre para inserir o novo valor.
        # ----------------------------------------------------
        while True:

            # ------------------------------------------------
            # Exibe o nó atual analisado.
            # ------------------------------------------------
            print(f"\nNó atual: {atual}")

            # ------------------------------------------------
            # Caso o novo valor seja MENOR que o nó atual:
            # deve seguir para a subárvore esquerda.
            # ------------------------------------------------
            if valor < atual:

                # --------------------------------------------
                # Exibe a decisão lógica.
                # --------------------------------------------
                print(f"{valor} < {atual}")
                print("Mover para ESQUERDA")

                # --------------------------------------------
                # Verifica se o filho esquerdo está vazio.
                # --------------------------------------------
                if arvore[atual]["esquerda"] is None:

                    # ----------------------------------------
                    # Exibe posição encontrada.
                    # ----------------------------------------
                    print(f"Posição esquerda de {atual} está livre.")

                    # ----------------------------------------
                    # Insere o novo valor à esquerda.
                    # ----------------------------------------
                    arvore[atual]["esquerda"] = valor

                    # ----------------------------------------
                    # Cria o novo nó no dicionário.
                    # ----------------------------------------
                    arvore[valor] = {
                        "esquerda": None,
                        "direita": None
                    }

                    # ----------------------------------------
                    # Exibe confirmação da inserção.
                    # ----------------------------------------
                    print(f"Valor {valor} inserido à ESQUERDA de {atual}")

                    # ----------------------------------------
                    # Finaliza o loop de inserção.
                    # ----------------------------------------
                    break

                else:

                    # ----------------------------------------
                    # Exibe continuidade da navegação.
                    # ----------------------------------------
                    print(
                        f"Filho esquerdo encontrado: "
                        f"{arvore[atual]['esquerda']}"
                    )

                    # ----------------------------------------
                    # Caso já exista filho esquerdo,
                    # continua navegando para baixo.
                    # ----------------------------------------
                    atual = arvore[atual]["esquerda"]

            else:

                # ------------------------------------------------
                # Caso o novo valor seja MAIOR OU IGUAL:
                # segue para a subárvore direita.
                # ------------------------------------------------

                # --------------------------------------------
                # Exibe decisão lógica.
                # --------------------------------------------
                print(f"{valor} >= {atual}")
                print("Mover para DIREITA")

                # --------------------------------------------
                # Verifica se o filho direito está vazio.
                # --------------------------------------------
                if arvore[atual]["direita"] is None:

                    # ----------------------------------------
                    # Exibe posição encontrada.
                    # ----------------------------------------
                    print(f"Posição direita de {atual} está livre.")

                    # ----------------------------------------
                    # Insere o valor à direita.
                    # ----------------------------------------
                    arvore[atual]["direita"] = valor

                    # ----------------------------------------
                    # Cria o novo nó.
                    # ----------------------------------------
                    arvore[valor] = {
                        "esquerda": None,
                        "direita": None
                    }

                    # ----------------------------------------
                    # Exibe confirmação da inserção.
                    # ----------------------------------------
                    print(f"Valor {valor} inserido à DIREITA de {atual}")

                    # ----------------------------------------
                    # Finaliza o processo de inserção.
                    # ----------------------------------------
                    break

                else:

                    # ----------------------------------------
                    # Exibe continuidade da navegação.
                    # ----------------------------------------
                    print(
                        f"Filho direito encontrado: "
                        f"{arvore[atual]['direita']}"
                    )

                    # ----------------------------------------
                    # Continua navegando para a direita.
                    # ----------------------------------------
                    atual = arvore[atual]["direita"]

    # --------------------------------------------------------
    # Exibe o estado parcial da árvore após cada inserção.
    # --------------------------------------------------------
    print("\nEstado atual da árvore:")
    print(arvore)


# ============================================================
# EXIBE A ÁRVORE FINAL
# ============================================================

print("\n================================================")
print("ÁRVORE FINAL")
print("================================================")
print(arvore)
```

```python
import matplotlib.pyplot as plt
import copy


# ============================================================
# FUNÇÃO PARA CALCULAR AS POSIÇÕES DOS NÓS NO GRÁFICO
# ============================================================

def calcular_posicoes(arvore, no, x=0, y=0, dx=1.0, posicoes=None):

    # --------------------------------------------------------
    # Cria o dicionário de posições caso ainda não exista.
    # --------------------------------------------------------
    if posicoes is None:
        posicoes = {}

    # --------------------------------------------------------
    # Processa apenas nós existentes.
    # --------------------------------------------------------
    if no is not None:

        # ----------------------------------------------------
        # Armazena a posição do nó atual.
        # ----------------------------------------------------
        posicoes[no] = (x, y)

        # ----------------------------------------------------
        # Calcula posições da subárvore esquerda.
        # ----------------------------------------------------
        calcular_posicoes(
            arvore,
            arvore[no]["esquerda"],
            x - dx,
            y - 1,
            dx / 2,
            posicoes
        )

        # ----------------------------------------------------
        # Calcula posições da subárvore direita.
        # ----------------------------------------------------
        calcular_posicoes(
            arvore,
            arvore[no]["direita"],
            x + dx,
            y - 1,
            dx / 2,
            posicoes
        )

    return posicoes
```

```python
# ============================================================
# FUNÇÃO PARA DESENHAR A ÁRVORE BINÁRIA
# ============================================================

def desenhar_arvore(
    arvore,
    raiz,
    titulo="Árvore Binária de Busca (BST)",
    destacado=None,
    caminho=None
):

    # --------------------------------------------------------
    # Verifica se a árvore está vazia.
    # --------------------------------------------------------
    if raiz is None or len(arvore) == 0:
        print("Árvore vazia.")
        return

    # --------------------------------------------------------
    # Se nenhum caminho for informado,
    # cria uma lista vazia.
    # --------------------------------------------------------
    if caminho is None:
        caminho = []

    # --------------------------------------------------------
    # Calcula as posições dos nós.
    # --------------------------------------------------------
    posicoes = calcular_posicoes(arvore, raiz)

    # --------------------------------------------------------
    # Cria a figura.
    # --------------------------------------------------------
    plt.figure(figsize=(12, 7))

    # ========================================================
    # DESENHA TODAS AS ARESTAS DA ÁRVORE
    # ========================================================

    for no, filhos in arvore.items():

        x1, y1 = posicoes[no]

        for lado in ["esquerda", "direita"]:

            filho = filhos[lado]

            if filho is not None:

                x2, y2 = posicoes[filho]

                plt.plot(
                    [x1, x2],
                    [y1, y2],
                    color="black",
                    linewidth=2,
                    zorder=1
                )

    # ========================================================
    # DESTACA O CAMINHO, QUANDO INFORMADO
    # ========================================================

    if len(caminho) > 1:

        for i in range(len(caminho) - 1):

            no1 = caminho[i]
            no2 = caminho[i + 1]

            x1, y1 = posicoes[no1]
            x2, y2 = posicoes[no2]

            plt.plot(
                [x1, x2],
                [y1, y2],
                color="red",
                linewidth=4,
                zorder=2
            )

    # ========================================================
    # DESENHA TODOS OS NÓS
    # ========================================================

    for no, (x, y) in posicoes.items():

        # ----------------------------------------------------
        # Define a cor do nó.
        # ----------------------------------------------------
        if no == destacado:
            facecolor = "lightcoral"

        elif no in caminho:
            facecolor = "lightgreen"

        else:
            facecolor = "white"

        plt.text(
            x,
            y,
            str(no),
            ha="center",
            va="center",
            fontsize=11,
            fontweight="bold",
            bbox=dict(
                boxstyle="circle",
                facecolor=facecolor,
                edgecolor="black",
                linewidth=2
            ),
            zorder=3
        )

    # ========================================================
    # CONFIGURAÇÕES FINAIS
    # ========================================================

    plt.title(titulo, fontsize=14)
    plt.axis("off")
    plt.show()
```

```python
desenhar_arvore(arvore, raiz, 'aaa')
```

## Inserção de novo valor na árvore binária de busca

A inserção de um novo valor na Árvore Binária de Busca já consolidada segue a mesma lógica de criação da árvore e consiste em localizar a posição correta para um novo valor, preservando a regra fundamental da estrutura: valores menores que um nó são inseridos à esquerda, enquanto valores maiores ou iguais são inseridos à direita.

```python
# ============================================================
# INSERÇÃO DE UM NOVO VALOR NA ÁRVORE BINÁRIA DE BUSCA
# ============================================================

# ------------------------------------------------------------
# Valor que será inserido na árvore.
# Altere esse valor para testar outras inserções.
# ------------------------------------------------------------
valor_inserir = 65


# ------------------------------------------------------------
# Verifica se a árvore está vazia.
# Se estiver vazia, o novo valor será a raiz.
# ------------------------------------------------------------
if len(arvore) == 0:

    # --------------------------------------------------------
    # Cria o nó raiz.
    # --------------------------------------------------------
    arvore[valor_inserir] = {
        "esquerda": None,
        "direita": None
    }

    # --------------------------------------------------------
    # Define o novo valor como raiz.
    # --------------------------------------------------------
    raiz = valor_inserir

    print(f"O valor {valor_inserir} foi inserido como raiz da árvore.")


else:

    # --------------------------------------------------------
    # Inicia a busca pela posição de inserção a partir da raiz.
    # --------------------------------------------------------
    atual = raiz

    # --------------------------------------------------------
    # Lista para armazenar o caminho percorrido até a inserção.
    # --------------------------------------------------------
    caminho_insercao = []

    # --------------------------------------------------------
    # Continua até encontrar uma posição vazia.
    # --------------------------------------------------------
    while True:

        # ----------------------------------------------------
        # Registra o nó atualmente analisado.
        # ----------------------------------------------------
        caminho_insercao.append(atual)

        print(f"\nNó atual: {atual}")

        # ----------------------------------------------------
        # Se o valor a inserir for menor que o nó atual,
        # a busca segue para a esquerda.
        # ----------------------------------------------------
        if valor_inserir < atual:

            print(f"{valor_inserir} < {atual}")
            print("Mover para ESQUERDA")

            # ------------------------------------------------
            # Se não houver filho esquerdo,
            # a posição correta foi encontrada.
            # ------------------------------------------------
            if arvore[atual]["esquerda"] is None:

                # --------------------------------------------
                # Conecta o novo valor como filho esquerdo.
                # --------------------------------------------
                arvore[atual]["esquerda"] = valor_inserir

                # --------------------------------------------
                # Cria o novo nó no dicionário.
                # --------------------------------------------
                arvore[valor_inserir] = {
                    "esquerda": None,
                    "direita": None
                }

                print(f"Valor {valor_inserir} inserido à ESQUERDA de {atual}.")
                break

            # ------------------------------------------------
            # Caso já exista filho esquerdo,
            # continua a navegação.
            # ------------------------------------------------
            else:
                atual = arvore[atual]["esquerda"]

        # ----------------------------------------------------
        # Se o valor a inserir for maior ou igual ao nó atual,
        # a busca segue para a direita.
        # ----------------------------------------------------
        else:

            print(f"{valor_inserir} >= {atual}")
            print("Mover para DIREITA")

            # ------------------------------------------------
            # Se não houver filho direito,
            # a posição correta foi encontrada.
            # ------------------------------------------------
            if arvore[atual]["direita"] is None:

                # --------------------------------------------
                # Conecta o novo valor como filho direito.
                # --------------------------------------------
                arvore[atual]["direita"] = valor_inserir

                # --------------------------------------------
                # Cria o novo nó no dicionário.
                # --------------------------------------------
                arvore[valor_inserir] = {
                    "esquerda": None,
                    "direita": None
                }

                print(f"Valor {valor_inserir} inserido à DIREITA de {atual}.")
                break

            # ------------------------------------------------
            # Caso já exista filho direito,
            # continua a navegação.
            # ------------------------------------------------
            else:
                atual = arvore[atual]["direita"]


# ============================================================
# RESULTADO DA INSERÇÃO
# ============================================================

print("\nCaminho percorrido na inserção:")
print(caminho_insercao)

print("\nÁrvore após a inserção:")
print(arvore)


# ============================================================
# VISUALIZAÇÃO GRÁFICA DA ÁRVORE APÓS A INSERÇÃO
# ============================================================

desenhar_arvore(
    arvore,
    raiz,
    "Árvore após a inserção",
    destacado=valor_inserir
)
```

## Busca na árvore binária de busca

```python
# ------------------------------------------------------------
# Valor que será procurado na árvore binária.
# ------------------------------------------------------------
valor_procurado = 49


# ------------------------------------------------------------
# Inicializa a busca a partir da raiz da árvore.
#
# A variável "atual" representa o nó atualmente analisado.
# ------------------------------------------------------------
atual = raiz


# ------------------------------------------------------------
# Lista utilizada para armazenar o caminho percorrido
# durante a busca.
#
# Isso permite visualizar quais nós foram visitados.
# ------------------------------------------------------------
caminho = []


# ------------------------------------------------------------
# Variável lógica que indica se o valor foi encontrado.
#
# False -> ainda não encontrado
# True  -> encontrado
# ------------------------------------------------------------
encontrou = False


# ============================================================
# PROCESSO DE BUSCA NA ÁRVORE BINÁRIA
# ============================================================

# ------------------------------------------------------------
# Continua a busca enquanto existir um nó válido.
#
# Quando "atual" se tornar None,
# significa que o valor não existe na árvore.
# ------------------------------------------------------------
while atual is not None:

    # --------------------------------------------------------
    # Adiciona o nó atual ao caminho percorrido.
    # --------------------------------------------------------
    caminho.append(atual)

    # --------------------------------------------------------
    # Verifica se o valor procurado é igual ao nó atual.
    # --------------------------------------------------------
    if valor_procurado == atual:

        # ----------------------------------------------------
        # Marca que o valor foi encontrado.
        # ----------------------------------------------------
        encontrou = True

        # ----------------------------------------------------
        # Interrompe o loop de busca.
        # ----------------------------------------------------
        break

    # --------------------------------------------------------
    # Caso o valor procurado seja menor que o nó atual,
    # a busca continua na subárvore esquerda.
    # --------------------------------------------------------
    elif valor_procurado < atual:

        # ----------------------------------------------------
        # Move para o filho esquerdo.
        # ----------------------------------------------------
        atual = arvore[atual]["esquerda"]

    # --------------------------------------------------------
    # Caso contrário,
    # o valor procurado é maior que o nó atual.
    #
    # Portanto, a busca continua à direita.
    # --------------------------------------------------------
    else:

        # ----------------------------------------------------
        # Move para o filho direito.
        # ----------------------------------------------------
        atual = arvore[atual]["direita"]


# ============================================================
# RESULTADO DA BUSCA
# ============================================================

# ------------------------------------------------------------
# Verifica se o valor foi encontrado.
# ------------------------------------------------------------
if encontrou:

    # --------------------------------------------------------
    # Exibe mensagem de sucesso.
    # --------------------------------------------------------
    print("Valor encontrado!")

else:

    # --------------------------------------------------------
    # Exibe mensagem de falha.
    # --------------------------------------------------------
    print("Valor não encontrado.")


# ------------------------------------------------------------
# Exibe o caminho percorrido durante a busca.
# ------------------------------------------------------------
print("Caminho percorrido:", caminho)
```

```python
# Plot destacando caminho de busca
desenhar_arvore(
    arvore,
    raiz,
    titulo="Busca em Árvore Binária de Busca (BST)",
    caminho=caminho

)
```

## Travessia pré-ordem

A travessia em pré-ordem é um método de percorrer os nós de uma árvore binária seguindo a sequência:

#### Raiz → Esquerda → Direita

Nesse tipo de travessia, o nó atual é visitado primeiro. Em seguida, percorre-se toda a subárvore esquerda e, depois, toda a subárvore direita.

Esse método é amplamente utilizado em:

- cópia de árvores;
- representação hierárquica;
- serialização de estruturas;
- expressão de operações prefixadas.

#### A lógica consiste em sempre processar o nó antes de seus filhos.

```python
# ============================================================
# TRAVESSIA EM PRÉ-ORDEM (PREORDER TRAVERSAL)
# ============================================================

# ------------------------------------------------------------
# Lista que armazenará o resultado da travessia.
#
# A travessia em pré-ordem segue a sequência:
#
# 1. Nó atual
# 2. Subárvore esquerda
# 3. Subárvore direita
# ------------------------------------------------------------
pre_ordem = []


# ------------------------------------------------------------
# Cria uma pilha contendo inicialmente a raiz da árvore.
#
# A pilha será utilizada para controlar os nós
# que ainda precisam ser visitados.
#
# Estrutura LIFO:
# Last In, First Out
# ------------------------------------------------------------
pilha = [raiz]


# ============================================================
# PROCESSO DE TRAVESSIA
# ============================================================

# ------------------------------------------------------------
# Continua enquanto existirem elementos na pilha.
# ------------------------------------------------------------
while len(pilha) > 0:

    # --------------------------------------------------------
    # Remove o último elemento da pilha.
    #
    # pop() remove sempre o elemento do topo.
    # --------------------------------------------------------
    atual = pilha.pop()

    # --------------------------------------------------------
    # Adiciona o nó atual na lista de saída.
    #
    # Como é pré-ordem,
    # o nó é processado imediatamente.
    # --------------------------------------------------------
    pre_ordem.append(atual)

    # --------------------------------------------------------
    # Verifica se existe filho à direita.
    # --------------------------------------------------------
    if arvore[atual]["direita"] is not None:

        # ----------------------------------------------------
        # Adiciona o filho direito na pilha.
        # ----------------------------------------------------
        pilha.append(arvore[atual]["direita"])

    # --------------------------------------------------------
    # Verifica se existe filho à esquerda.
    # --------------------------------------------------------
    if arvore[atual]["esquerda"] is not None:

        # ----------------------------------------------------
        # Adiciona o filho esquerdo na pilha.
        # ----------------------------------------------------
        pilha.append(arvore[atual]["esquerda"])

    # --------------------------------------------------------
    # IMPORTANTE:
    #
    # O filho direito é inserido PRIMEIRO.
    # O filho esquerdo é inserido DEPOIS.
    #
    # Como a pilha funciona em LIFO,
    # o filho esquerdo será processado antes.
    #
    # Isso mantém a ordem correta:
    #
    # Nó -> Esquerda -> Direita
    # --------------------------------------------------------


# ============================================================
# EXIBE O RESULTADO FINAL
# ============================================================

print("Travessia pré-ordem:")
print(pre_ordem)
```

## Travessia em ordem

A travessia em ordem, também chamada de in-order traversal, é um método de percorrer os nós de uma árvore binária seguindo a sequência:

#### Esquerda → Raiz → Direita

Nesse tipo de travessia, o algoritmo primeiro percorre toda a subárvore esquerda. Depois visita o nó atual (raiz) e, por fim, percorre a subárvore direita.

Em árvores binárias de busca (BST), essa travessia possui uma propriedade importante:

Os valores são visitados em ordem crescente.

A travessia em ordem é amplamente utilizada para:

- ordenação de elementos em BST;
- recuperação de dados em sequência crescente;
- geração de listas ordenadas;
- processamento hierárquico estruturado.

```python
# ============================================================
# TRAVESSIA EM ORDEM (IN-ORDER TRAVERSAL)
# ============================================================

# ------------------------------------------------------------
# Lista que armazenará o resultado final da travessia.
#
# A travessia em ordem segue a sequência:
#
# 1. Subárvore esquerda
# 2. Nó atual (raiz)
# 3. Subárvore direita
# ------------------------------------------------------------
em_ordem = []


# ------------------------------------------------------------
# Pilha utilizada para controlar os nós que ainda precisam ser processados.
# A pilha simula o comportamento da recursão.
# ------------------------------------------------------------
pilha = []


# ------------------------------------------------------------
# Inicializa a navegação a partir da raiz da árvore.
# ------------------------------------------------------------
atual = raiz


# ============================================================
# PROCESSO DE TRAVESSIA
# ============================================================

# ------------------------------------------------------------
# O loop continua enquanto:
#
# 1. existir um nó atual válido
# OU
# 2. ainda existirem nós armazenados na pilha
# ------------------------------------------------------------
while atual is not None or len(pilha) > 0:

    # ========================================================
    # DESCIDA PELA ESQUERDA
    # ========================================================

    # --------------------------------------------------------
    # Continua descendo para a esquerda
    # até encontrar um nó sem filho esquerdo.
    # --------------------------------------------------------
    while atual is not None:

        # ----------------------------------------------------
        # Armazena o nó atual na pilha.
        #
        # Isso permite retornar posteriormente
        # para processar a raiz e seguir à direita.
        # ----------------------------------------------------
        pilha.append(atual)

        # ----------------------------------------------------
        # Move para o filho esquerdo.
        # ----------------------------------------------------
        atual = arvore[atual]["esquerda"]

    # ========================================================
    # PROCESSAMENTO DO NÓ
    # ========================================================

    # --------------------------------------------------------
    # Remove o último nó armazenado na pilha.
    #
    # Esse nó representa:
    #
    # - o nó mais à esquerda ainda não processado.
    # --------------------------------------------------------
    atual = pilha.pop()

    # --------------------------------------------------------
    # Adiciona o nó atual no resultado da travessia.
    # --------------------------------------------------------
    em_ordem.append(atual)

    # ========================================================
    # NAVEGAÇÃO PARA A DIREITA
    # ========================================================

    # --------------------------------------------------------
    # Após processar o nó atual,
    # a travessia continua na subárvore direita.
    # --------------------------------------------------------
    atual = arvore[atual]["direita"]


# ============================================================
# EXIBE O RESULTADO FINAL
# ============================================================

print("Travessia em ordem:")
print(em_ordem)
```

## Travessia pós-ordem

A travessia pós-ordem, também chamada de post-order traversal, é um método de percorrer uma árvore binária seguindo a sequência:

#### Esquerda → Direita → Raiz

Nesse tipo de travessia, o algoritmo primeiro percorre toda a subárvore esquerda, depois toda a subárvore direita e, somente ao final, processa o nó atual.

A principal característica dessa travessia é que a raiz é visitada por último.

A pós-ordem é frequentemente utilizada em:

- remoção de árvores;
- liberação de memória;
- avaliação de expressões;
- processamento hierárquico dependente dos filhos.

```python
# ============================================================
# TRAVESSIA PÓS-ORDEM (POST-ORDER TRAVERSAL)
# ============================================================

# ------------------------------------------------------------
# Lista que armazenará o resultado final da travessia.
#
# A pós-ordem segue a sequência:
#
# 1. Subárvore esquerda
# 2. Subárvore direita
# 3. Raiz
# ------------------------------------------------------------
pos_ordem = []


# ------------------------------------------------------------
# Primeira pilha utilizada para percorrer a árvore.
#
# Inicialmente contém apenas a raiz.
# ------------------------------------------------------------
pilha1 = [raiz]


# ------------------------------------------------------------
# Segunda pilha utilizada para inverter a ordem de visita.
#
# Essa inversão permitirá obter:
#
# esquerda → direita → raiz
# ------------------------------------------------------------
pilha2 = []


# ============================================================
# PRIMEIRA ETAPA DA TRAVESSIA
# ============================================================

# ------------------------------------------------------------
# Continua enquanto existirem elementos na primeira pilha.
# ------------------------------------------------------------
while len(pilha1) > 0:

    # --------------------------------------------------------
    # Remove o elemento do topo da pilha1.
    # --------------------------------------------------------
    atual = pilha1.pop()

    # --------------------------------------------------------
    # Adiciona o nó atual na pilha2.
    #
    # A pilha2 armazenará os nós em ordem invertida.
    # --------------------------------------------------------
    pilha2.append(atual)

    # --------------------------------------------------------
    # Verifica se existe filho à esquerda.
    # --------------------------------------------------------
    if arvore[atual]["esquerda"] is not None:

        # ----------------------------------------------------
        # Adiciona o filho esquerdo na pilha1.
        # ----------------------------------------------------
        pilha1.append(arvore[atual]["esquerda"])

    # --------------------------------------------------------
    # Verifica se existe filho à direita.
    # --------------------------------------------------------
    if arvore[atual]["direita"] is not None:

        # ----------------------------------------------------
        # Adiciona o filho direito na pilha1.
        # ----------------------------------------------------
        pilha1.append(arvore[atual]["direita"])


# ============================================================
# SEGUNDA ETAPA DA TRAVESSIA
# ============================================================

# ------------------------------------------------------------
# Remove os elementos da pilha2.
#
# Como a pilha é LIFO,
# os elementos sairão na ordem correta da pós-ordem:
#
# esquerda → direita → raiz
# ------------------------------------------------------------
while len(pilha2) > 0:

    # --------------------------------------------------------
    # Remove o topo da pilha2
    # e adiciona no resultado final.
    # --------------------------------------------------------
    pos_ordem.append(pilha2.pop())


# ============================================================
# EXIBE O RESULTADO FINAL
# ============================================================

print("Travessia pós-ordem:")
print(pos_ordem)
```

## Exclusão na árvore binária de busca

A exclusão em uma Árvore Binária de Busca (BST) é uma operação mais complexa do que inserção e busca, pois a remoção de um nó pode alterar a estrutura da árvore e comprometer sua propriedade de ordenação.

Em uma BST, todos os valores menores ficam à esquerda e todos os maiores ficam à direita. Durante a exclusão, essa regra precisa ser preservada.

Existem três casos principais:

1. Nó folha
2. Nó com um filho
3. Nó com dois filhos

Os dois primeiros casos são relativamente simples. O maior desafio ocorre quando o nó possui dois filhos.

Nesse cenário, o nó removido não pode simplesmente desaparecer, pois isso quebraria a organização da árvore. Assim, o algoritmo substitui o valor removido pelo sucessor em ordem.

#### O sucessor em ordem é o menor valor da subárvore direita.

A lógica geral da exclusão é:

1. Localizar o nó
2. Identificar o caso de remoção
3. Reorganizar os ponteiros
4. Preservar a propriedade da BST

```python



# ============================================================
# FUNÇÃO PARA DESENHAR A ÁRVORE
# ============================================================

def desenhar_arvore(arvore, raiz, titulo, destacado=None):

    if raiz is None:
        print("Árvore vazia.")
        return

    posicoes = calcular_posicoes(arvore, raiz)

    plt.figure(figsize=(10, 6))

    # --------------------------------------------------------
    # Desenha as arestas.
    # --------------------------------------------------------
    for no, filhos in arvore.items():

        x1, y1 = posicoes[no]

        for lado in ["esquerda", "direita"]:

            filho = filhos[lado]

            if filho is not None:
                x2, y2 = posicoes[filho]

                plt.plot(
                    [x1, x2],
                    [y1, y2],
                    color="black",
                    linewidth=2,
                    zorder=1
                )

    # --------------------------------------------------------
    # Desenha os nós.
    # --------------------------------------------------------
    for no, (x, y) in posicoes.items():

        if no == destacado:
            cor = "lightcoral"
        else:
            cor = "white"

        plt.text(
            x,
            y,
            str(no),
            ha="center",
            va="center",
            fontsize=11,
            fontweight="bold",
            bbox=dict(
                boxstyle="circle",
                facecolor=cor,
                edgecolor="black",
                linewidth=2
            ),
            zorder=2
        )

    plt.title(titulo)
    plt.axis("off")
    plt.show()


# ============================================================
# REMOÇÃO EM ÁRVORE BINÁRIA DE BUSCA
# ============================================================

# ------------------------------------------------------------
# Valor que será removido da árvore.
# ------------------------------------------------------------
valor_remover = 72


# ------------------------------------------------------------
# Guarda uma cópia da árvore antes da remoção.
# Isso será usado para gerar o gráfico comparativo.
# ------------------------------------------------------------
arvore_antes = copy.deepcopy(arvore)
raiz_antes = raiz


# ------------------------------------------------------------
# Inicializa a busca pelo nó a ser removido.
# ------------------------------------------------------------
atual = raiz


# ------------------------------------------------------------
# Variável que armazenará o pai do nó atual.
# A raiz não possui pai, por isso começa como None.
# ------------------------------------------------------------
pai = None


# ============================================================
# PRIMEIRA ETAPA: LOCALIZAR O NÓ A SER REMOVIDO
# ============================================================

while atual is not None and atual != valor_remover:

    # --------------------------------------------------------
    # Antes de descer na árvore, guarda o nó atual como pai.
    # --------------------------------------------------------
    pai = atual

    # --------------------------------------------------------
    # Se o valor a remover for menor que o nó atual,
    # a busca segue pela esquerda.
    # --------------------------------------------------------
    if valor_remover < atual:
        atual = arvore[atual]["esquerda"]

    # --------------------------------------------------------
    # Caso contrário, segue pela direita.
    # --------------------------------------------------------
    else:
        atual = arvore[atual]["direita"]


# ============================================================
# SEGUNDA ETAPA: VERIFICAR SE O VALOR FOI ENCONTRADO
# ============================================================

if atual is None:

    print("Valor não encontrado para remoção.")

else:

    # --------------------------------------------------------
    # Guarda os filhos esquerdo e direito do nó encontrado.
    # --------------------------------------------------------
    esquerda = arvore[atual]["esquerda"]
    direita = arvore[atual]["direita"]


    # ========================================================
    # CASO 1: NÓ FOLHA
    # ========================================================

    if esquerda is None and direita is None:

        # ----------------------------------------------------
        # Se o nó removido for a raiz e não tiver filhos,
        # a árvore fica vazia.
        # ----------------------------------------------------
        if pai is None:
            arvore = {}
            raiz = None

        # ----------------------------------------------------
        # Se o nó for filho esquerdo do pai,
        # remove a ligação à esquerda.
        # ----------------------------------------------------
        elif arvore[pai]["esquerda"] == atual:
            arvore[pai]["esquerda"] = None

        # ----------------------------------------------------
        # Caso contrário, remove a ligação à direita.
        # ----------------------------------------------------
        else:
            arvore[pai]["direita"] = None

        # ----------------------------------------------------
        # Remove o nó do dicionário.
        # ----------------------------------------------------
        del arvore[atual]


    # ========================================================
    # CASO 2: NÓ COM APENAS UM FILHO
    # ========================================================

    elif esquerda is None or direita is None:

        # ----------------------------------------------------
        # Identifica qual filho existe.
        # ----------------------------------------------------
        if esquerda is not None:
            filho = esquerda
        else:
            filho = direita

        # ----------------------------------------------------
        # Se o nó removido for a raiz,
        # o filho passa a ser a nova raiz.
        # ----------------------------------------------------
        if pai is None:
            raiz = filho

        # ----------------------------------------------------
        # Se o nó removido for filho esquerdo do pai,
        # o pai passa a apontar para o filho do nó removido.
        # ----------------------------------------------------
        elif arvore[pai]["esquerda"] == atual:
            arvore[pai]["esquerda"] = filho

        # ----------------------------------------------------
        # Caso contrário, atualiza o filho direito do pai.
        # ----------------------------------------------------
        else:
            arvore[pai]["direita"] = filho

        # ----------------------------------------------------
        # Remove o nó do dicionário.
        # ----------------------------------------------------
        del arvore[atual]


    # ========================================================
    # CASO 3: NÓ COM DOIS FILHOS
    # ========================================================

    else:

        # ----------------------------------------------------
        # Para remover um nó com dois filhos,
        # usa-se o sucessor em ordem.
        #
        # O sucessor é o menor valor da subárvore direita.
        # ----------------------------------------------------
        pai_sucessor = atual
        sucessor = arvore[atual]["direita"]

        # ----------------------------------------------------
        # Caminha para a esquerda dentro da subárvore direita
        # até encontrar o menor valor.
        # ----------------------------------------------------
        while arvore[sucessor]["esquerda"] is not None:
            pai_sucessor = sucessor
            sucessor = arvore[sucessor]["esquerda"]

        # ----------------------------------------------------
        # Guarda o filho direito do sucessor.
        # O sucessor não possui filho esquerdo,
        # pois ele é o menor da subárvore direita.
        # ----------------------------------------------------
        filho_direito_sucessor = arvore[sucessor]["direita"]

        # ----------------------------------------------------
        # Cria uma cópia dos filhos do nó removido.
        # O sucessor assumirá essa posição.
        # ----------------------------------------------------
        arvore[sucessor] = {
            "esquerda": arvore[atual]["esquerda"],
            "direita": arvore[atual]["direita"]
        }

        # ----------------------------------------------------
        # Se o sucessor era filho direto do nó removido,
        # ajusta sua subárvore direita.
        # ----------------------------------------------------
        if pai_sucessor == atual:
            arvore[sucessor]["direita"] = filho_direito_sucessor

        # ----------------------------------------------------
        # Caso o sucessor estivesse mais abaixo,
        # o pai do sucessor passa a apontar para o filho direito dele.
        # ----------------------------------------------------
        else:
            arvore[pai_sucessor]["esquerda"] = filho_direito_sucessor

        # ----------------------------------------------------
        # Atualiza a ligação do pai do nó removido.
        # ----------------------------------------------------
        if pai is None:
            raiz = sucessor

        elif arvore[pai]["esquerda"] == atual:
            arvore[pai]["esquerda"] = sucessor

        else:
            arvore[pai]["direita"] = sucessor

        # ----------------------------------------------------
        # Remove o nó original.
        # ----------------------------------------------------
        del arvore[atual]


    # ========================================================
    # RESULTADO FINAL
    # ========================================================

    print("Valor removido:", valor_remover)
    print("Árvore após remoção:")
    print(arvore)


# ============================================================
# PLOTS DA REMOÇÃO
# ============================================================

desenhar_arvore(
    arvore_antes,
    raiz_antes,
    "Árvore antes da remoção",
    destacado=valor_remover
)

desenhar_arvore(
    arvore,
    raiz,
    "Árvore após a remoção"
)
```
