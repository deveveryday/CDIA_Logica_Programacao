### 2. Gráficos Multiplos e Subplot
#### Gráficos em uma mesma figura
- Com a função subplots (), você pode desenhar vários gráficos em uma figura:
- Nesse primeiro exemplo, vamos plotar dois sub gráficos de linhas

````python
plt.figure(figsize = (11, 6))

plt.subplot(1, 2, 1)
plt.title("Position Graph 01", fontsize = 16)

plt.subplot(1, 2, 2)
plt.title("Position Graph 02", fontsize = 16)

plt.suptitle("Graphics")

plt.show()
````

### Gráficos lado a lado

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(1, 2, 1)
plt.title("Position Graph 01", fontsize = 16)
plt.plot(x_v1, y_v1)

x_v1 = np.arange(0, 10, 0.2)
y_v1 = np.sin(x_v1)

plt.subplot(1, 2, 2)
plt.title("Position Graph 02", fontsize = 16)
plt.plot(x_v1, y_v1)

plt.suptitle("Graphics")

plt.show()
````

### Gráfico um em cima do outro

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(2, 1, 1)
plt.plot(x_v1, y_v1)

arr = []
arr2 = []
for i in range(100):
  arr.append(i)

for i in range(100):
  arr2.append(i)


x_v1 = np.array(arr)
y_v1 = np.array(arr2)

plt.subplot(2, 1, 2)
plt.plot(x_v1, y_v1)

plt.suptitle("Graphics")

plt.show()
````

### Múltiplos Gráficos com Subplot
- É possível plotar quantos gráficos quiser em uma figura, só precisar informar a quantidade de linhas e colunas que terá a figura, além do índice do gráfico.

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(2, 2, 1)
plt.plot(x_v1, y_v1)

arr = []
arr2 = []
arr3 = []

for i in range(100, 200):
  arr.append(i)

for i in range(100):
  arr2.append(i)

for i in range(100):
  arr3.append(i)


x_v1 = np.array(arr)
y_v1 = np.array(arr2)
y_v2 = np.array(arr3)

plt.subplot(2, 2, 2)
plt.plot(x_v1, y_v1)

plt.subplot(2, 2, 3)
plt.plot(x_v1, y_v1)

plt.subplot(2, 2, 4)
plt.plot(x_v1, np.log(y_v2))


plt.suptitle("Graphics")

plt.show()
````

### 4. Gráficos boxplot com Matplotlib e Seaborn
- Exemplos de boxplot, com a base de dados flores iris, que tem 3 tipos de flores

````python
data_base = sns.load_dataset("iris")
````

````python
plt.figure(figsize = (12, 6))

sns.boxplot(x = "species", y = "petal_length", hue = "species", data = data_base, width = 0.1, palette = ["red", "purple", "blue"])
#sns.boxplot(x = "species", y = "petal_length", hue = "species", data = data_base, width = 0.9, palette = "viridis")

plt.show()
````

### Mult Boxplot com Matplotlib

````python
setosa_data  = data_base[data_base["species"] == "setosa"]["petal_length"]
versicolor_data  = data_base[data_base["species"] == "versicolor"]["petal_length"]
virginica_data  = data_base[data_base["species"] == "virginica"]["petal_length"]

data = [setosa_data, versicolor_data, virginica_data]
````

````python
plt.figure(figsize = (8, 6))
box = plt.boxplot(data, patch_artist = True, tick_labels = ["setosa", "versicolor", "virginica"])

for patch, color in zip(box["boxes"], ["blue", "pink", "orange"]):
    patch.set_facecolor(color)
````

#### Formatando por partes

````python
plt.figure(figsize = (8, 6))
ticks = ["setosa", "versicolor", "virginica"]
colors = ["gray", "red", "green"]
whisker_colors = ["pink", "pink", "yellow", "yellow", "blue", "blue"]
cap_colors = ["blue", "blue", "yellow", "yellow", "blue", "blue"]

box = plt.boxplot(data, patch_artist = False, tick_labels = ticks)

for i, bplot in enumerate(box["boxes"]):
  bplot.set(color = colors[i], linewidth=3)

#whiskers
for i, whisker in enumerate(box["whiskers"]):
  whisker.set(color = whisker_colors[i], linewidth=3)

#caps
for i, cap in enumerate(box["caps"]):
  cap.set(color = cap_colors[i], linewidth=3)


#median
for i, median in enumerate(box["medians"]):
  median.set(color = colors[i], linewidth=3)


plt.show()
````

### Intervalos de modelagem de séries temporais
- A bibliteca tsmoothie serve para fazer a suavização de séries temporais e detecção de outliers de forma vetorizada.

````python
#!pip install tsmoothie
````

````python
x = np.linspace(0,2 * np.pi, 100)
y = np.sin(x) + np.random.random(100) * 0.2 #geração do sinal: seno + ruído uniforme

plt.figure(figsize = (11, 6))
plt.plot(y, ".y")
plt.show()
````

````python
x = np.linspace(0,2 * np.pi, 100)
y = np.sin(x) + np.random.random(100) * 0.3 #geração do sinal: seno + ruído uniforme

# Etapa de suavização

smoother = ConvolutionSmoother(window_len = 5, window_type = "ones")
smoother.smooth(y) # aplica suavização no y

low, up = smoother.get_intervals(
    "sigma_interval", # tipo de intervalo (baseado no desvio padrão)
    n_sigma = 2 # intervalo de +ou- 2 desvios padrão sob a normalidade
)

#386 rows  - 95 confidance - 20mi


plt.figure(figsize = (11, 6))
plt.plot(smoother.smooth_data[0], linewidth = 3, color = "red")
plt.plot(y, ".y")


plt.fill_between(range(len(smoother.data[0])),
                 low[0],
                 up[0],
                 alpha = 0.3
                )
plt.show()
````
