## 1. Gráfico de linhas
##### Sintaxe básica:
``` Python
    plt.plot(x, y)

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
````

````python
x = []
y = []

for i in range(1, 1000, 2):
  x.append(i)

for i in range(10, 1010, 2):
  y.append(i)
````

````python
plt.plot(x, y)
plt.title("Title")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
import numpy as np

start = 0
stop = 10
step = 0.1

new_x = np.arange(start, stop, step)
new_y = np.sin(new_x)

plt.plot(new_x, new_y)
plt.show()
````

#### Gráfico de linhas (adicionando elementos)

````python
color = "purple"
label = "Seno"
line_width = 5

plt.figure(figsize = (15, 6), dpi = 150)
plt.plot(new_x, new_y, color = color, label = label, linewidth = line_width)
plt.title("Graph Sci", fontsize = 17)
plt.legend(loc = "upper center")
plt.show()
````

### Gráficos lado a lado

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(1, 2, 1)
plt.title("Position Graph 01", fontsize = 16)
plt.plot(x_v1, y_v1)

x_v1 = np.arange(0, 10, 0.2)
y_v1 = np.sin(x_v1)

plt.subplot(1, 2, 2)
plt.title("Position Graph 02", fontsize = 16)
plt.plot(x_v1, y_v1)

plt.suptitle("Graphics")

plt.show()
````

### Gráfico um em cima do outro

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(2, 1, 1)
plt.plot(x_v1, y_v1)

arr = []
arr2 = []
for i in range(100):
  arr.append(i)

for i in range(100):
  arr2.append(i)


x_v1 = np.array(arr)
y_v1 = np.array(arr2)

plt.subplot(2, 1, 2)
plt.plot(x_v1, y_v1)

plt.suptitle("Graphics")

plt.show()
````

### Múltiplos Gráficos com Subplot
- É possível plotar quantos gráficos quiser em uma figura, só precisar informar a quantidade de linhas e colunas que terá a figura, além do índice do gráfico.

````python
plt.figure(figsize = (11, 6))

x_v1 = np.arange(0, 10, 0.1)
y_v1 = np.cosh(x_v1)

plt.subplot(2, 2, 1)
plt.plot(x_v1, y_v1)

arr = []
arr2 = []
arr3 = []

for i in range(100, 200):
  arr.append(i)

for i in range(100):
  arr2.append(i)

for i in range(100):
  arr3.append(i)


x_v1 = np.array(arr)
y_v1 = np.array(arr2)
y_v2 = np.array(arr3)

plt.subplot(2, 2, 2)
plt.plot(x_v1, y_v1)

plt.subplot(2, 2, 3)
plt.plot(x_v1, y_v1)

plt.subplot(2, 2, 4)
plt.plot(x_v1, np.log(y_v2))


plt.suptitle("Graphics")

plt.show()
````

### 4. Gráficos boxplot com Matplotlib e Seaborn
- Exemplos de boxplot, com a base de dados flores iris, que tem 3 tipos de flores

````python
data_base = sns.load_dataset("iris")
````

````python
plt.figure(figsize = (12, 6))

sns.boxplot(x = "species", y = "petal_length", hue = "species", data = data_base, width = 0.1, palette = ["red", "purple", "blue"])
#sns.boxplot(x = "species", y = "petal_length", hue = "species", data = data_base, width = 0.9, palette = "viridis")

plt.show()
````

### Mult Boxplot com Matplotlib

````python
setosa_data  = data_base[data_base["species"] == "setosa"]["petal_length"]
versicolor_data  = data_base[data_base["species"] == "versicolor"]["petal_length"]
virginica_data  = data_base[data_base["species"] == "virginica"]["petal_length"]

data = [setosa_data, versicolor_data, virginica_data]
````

````python
plt.figure(figsize = (8, 6))
box = plt.boxplot(data, patch_artist = True, tick_labels = ["setosa", "versicolor", "virginica"])

for patch, color in zip(box["boxes"], ["blue", "pink", "orange"]):
    patch.set_facecolor(color)
````

#### Formatando por partes

````python
plt.figure(figsize = (8, 6))
ticks = ["setosa", "versicolor", "virginica"]
colors = ["gray", "red", "green"]
whisker_colors = ["pink", "pink", "yellow", "yellow", "blue", "blue"]
cap_colors = ["blue", "blue", "yellow", "yellow", "blue", "blue"]

box = plt.boxplot(data, patch_artist = False, tick_labels = ticks)

for i, bplot in enumerate(box["boxes"]):
  bplot.set(color = colors[i], linewidth=3)

#whiskers
for i, whisker in enumerate(box["whiskers"]):
  whisker.set(color = whisker_colors[i], linewidth=3)

#caps
for i, cap in enumerate(box["caps"]):
  cap.set(color = cap_colors[i], linewidth=3)


#median
for i, median in enumerate(box["medians"]):
  median.set(color = colors[i], linewidth=3)


plt.show()
````

### Mult Boxplot com Matplotlib Cor

````python
plt.figure(figsize = (15, 6))
ticks = ["setosa", "versicolor", "virginica"]
colors = ["gray", "red", "green"]
whisker_colors = ["pink", "pink", "yellow", "yellow", "blue", "blue"]
cap_colors = ["blue", "blue", "yellow", "yellow", "blue", "blue"]

box = plt.boxplot(data, notch = True, patch_artist = True, vert = False, tick_labels = ticks)

colors = ["gray", "red", "green"]

for i, bplot in enumerate(box["boxes"]):
  bplot.set(color = "gray", linewidth=3)
  bplot.set(facecolor = colors[i])


for whiskers in box["whiskers"]:
  whiskers.set(color = "gray", linewidth = 3)


for cap in box["caps"]:
  cap.set(color = "gray", linewidth = 3)


for median in box["medians"]:
  median.set(color = "gray", linewidth = 20)

plt.show()
````

### 5. Visualização de Dados Categóricos
#### Gráfico de Barras

````python
plt.figure(figsize = (15, 9))
category = ["A", "B", "C", "D", "E"]
values = [44, 55, 66, 77, 88]

plt.subplot(1, 2, 1)
plt.bar(category, values, color = ["red", "blue", "yellow", "black", "green"])
plt.title("Position Graph 01", fontsize = 16)

plt.subplot(1, 2, 2)
plt.barh(category, values, color = ["red", "blue", "yellow", "black", "green"])
plt.title("Position Graph 02", fontsize = 16)

plt.show()
````

````python
df = pd.DataFrame({"Categories": category, "Values": values})

plt.figure(figsize = (15, 9))

sns.barplot(x = "Categories", y = "Values", data = df , hue = "Categories", palette = "viridis")


plt.show()
````

#### Gráfico de Torta ou Pizza

````python
plt.figure(figsize = (15, 9))

plt.pie(values, labels = category,
        autopct = "%1.1f%%", #show the percentual with 1 point
        startangle = 100, #rotates to start the first cut in the top
        colors = ["red", "blue", "yellow", "gray", "green"]
        )
plt.show()
````

#### Plotly (interativo)

````python
fig = px.pie(df, values = "Values", names = "Categories", title = "")
fig.show()
````

#### Heatmap (Matriz de Correlação)

````python
data = pd.DataFrame(np.random.randn(100, 8),
                    columns = ["a", "b", "c", "d", "e", "f", "g", "h"])
corr = data.corr()

plt.figure(figsize = (8, 6))
sns.heatmap(corr, annot = True, cmap = "cool", vmin = -1, vmax = 1)
#matriz de correlação
plt.show()
````

### 6. Visualização de Séries Temporais

````python
plt.figure(figsize = (12, 6))
#série temporal
dates = pd.date_range("2022-01-01", periods = 100)

#cumsum acumula dados com base no anterior
values = np.cumsum(np.random.randn(100))

plt.plot(dates, values, marker = "+", linestyle = "-")
plt.title("Title")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(rotation = 45)
plt.show()


````

#### Série temporal interativa com Plotly

````python
df = pd.DataFrame({"Data" : dates, "Value": values})

fig = px.line(df, x = "Data", y = "Value", title = "Série Temporal Interativa")
fig.update_xaxes(rangeslider_visible = True)
fig.show()
````
